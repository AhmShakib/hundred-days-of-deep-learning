```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("Placement.csv")
df.head()
```

```python
df.shape
```

```python

sns.scatterplot(x=df['cgpa'],y=df['resume_score'],hue=df['placed'])
```

```python
X = df[['cgpa', 'resume_score']]
y = df['placed']
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=20)
```

```python
pn=Perceptron()
```

```python
pn.fit(x_train,y_train)
```

```python
pn.coef_
```

```python
pn.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x_train.values, y_train.values, clf=pn, legend=2)
plt.xlabel("CGPA")
plt.ylabel("Resume Score")
plt.title("Perceptron Decision Boundary")
plt.show()
```

```python
y_pred = pn.predict(x_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
plot_decision_regions(x_test.values, y_test.values, clf=pn)
plt.title("Decision Boundary on Test Data")
plt.show()
```
